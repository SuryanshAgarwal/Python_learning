```python
def check_unique(data):
    if len(data) == len(set(data)):
        return True
    else:
        return False
```

```python
a = [1,4,3,2,5,3,2]
check_unique(a)
```

```python
def unique_triplets(uarr):
    temp = set()
    final_set = set()
    for i in uarr:
        for j in uarr:
            for k in uarr:
                temp.add((i,j,k))
    
    for triplets in temp:
        print(triplets)
        sum=triplets[0]+triplets[1]+triplets[2]
        if sum == 0:
            final_set.add(triplets)
    return final_set
```

```python
arr = [1,-2,4,2,-5,2,-4,10]
triplets = list(unique_triplets(arr))
```
